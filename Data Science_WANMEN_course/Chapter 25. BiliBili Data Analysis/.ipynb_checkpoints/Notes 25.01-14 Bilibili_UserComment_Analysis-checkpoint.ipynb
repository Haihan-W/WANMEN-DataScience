## 结巴分词

** 结合基于规则和基于统计 **
1. 基于前缀词典实现高效的词图扫描，生成句子中汉字所有可能成词情况所构成的有向无环图 (DAG)
2. 采用动态规划查找最大概率路径, 找出基于词频的最大切分组合

跳进未名湖

** 1.根据离线统计词典查找到的词频与词性 **

|词语|频率|词性|
| ------------- |:-------------:| -----:|
|跳| 8462 |v
|进| 25668 |v
|跳进| 206 |v
|未名湖| 27 |ns
|未名| 6 |v
|未| 16803 |d
|名| 26255 |q
|湖| 12466 |ns

** 2. 构建有向无环图 **

 0: [0,1]
 
 1: [1]
 
 2: [2,3,4]
 
 3: [3]
 
 4: [4]

### 去停用词


* 加载停用词词表


```python
# './stopword_chinese.txt' -> use this pre-defined dictionary, create a chinese stopword list
# 创建停用词list  
def create_stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf8').readlines()]  
    return stopwords  

stopwords = create_stopwordslist('./stopword_chinese.txt') #check "stopword_chinese.txt" for details (in current folder)
#stopwords.pop(0)
stopwords.append(' ')

# \ufeff IS THE bom, byte order mark in python
```

```python
print(stopwords)
```

* 去停用词

```python
refined_words = []
for word in danmu_fenci:
    if word not in stopwords:
        if word !='\t':
            refined_words.append(word)
```

```python
refined_words
```

### 整理词频

```python
#from collections import Counter -> A Counter is a container that keeps track of how many times equivalent values are added.
#https://pymotw.com/2/collections/counter.html
from collections import Counter
c = Counter(refined_words)
```

```python
c
```

#### sort by key

```python
#items-> key, value pair -> key=e.g. '背后','祖国',... from counter; value=e.g. 379,388,..., i.e. key's frequency
sorted_c_key = sorted(c.items(),key=lambda t:t[0]) #t[0] means first index of the item, i.e. key; t[1] means second index of item, i.e. value
sorted_c_key
```

#### sort by value

```python
sorted_c_value = sorted(c.items(),key=lambda t:t[1],reverse=True)
sorted_c_value
```

- IDF is calculated using jieba.analyse library

```python
from jieba import analyse
tuned_comments = " ".join(refined_words)
tags_tfidf=analyse.extract_tags(tuned_comments,topK=20)
```

```python
tags_tfidf
```

#compare this results to sorted_c_value (i.e. TF) above, we can find certain high frequency word was ranked lower, e.g. 祖国
#because its IDF value is low,  which means it is a common word in language library, word's 辨识度较低 

#### textrank算法进行关键词提取

** 算法 **
1. 识别文本单词，将其作为图的顶点置入图中
2. 一一计算文本**单词对**的关系（权重），并用这样的权重构造出图中对应的边
3. 基于textrank迭代计算公式，计算每个顶点（单词）的权重，直至收敛
4. 按照顶点的分数降序排列

```python
Image(filename='./textranking_form.jpg')
```

* textrank = analyse.textrank
* tags_textrank = textrank(comments,topK=20,withWeight=True,allowPOS=('ns','n','vn','v'))

```python
#put all filenames of csv into a file name list. (here I don't have all csv files in the folder, so the following is just a code demo)
filenames=[]
for i in range(12):
    name='./第一季第'+str(int(i+1))+'集.csv'
    filenames.append(name)
```

* 不同剧集的高频词

```python
def rm_stop(danmu_fenci,stopwords):
    refined_words = []
    for word in danmu_fenci:
        if word not in stopwords:
            if word !='\t':
                refined_words.append(word)
    return refined_words

# './stopword_chinese.txt'
# 创建停用词list  
def create_stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf8').readlines()]  
    return stopwords  

stopwords = create_stopwordslist('./stopword_chinese.txt')
#stopwords.pop(0)
stopwords.append(' ')
```

```python
HF_content = {}

for i in range(12):
    #the following is just a repeat of all of the content above
    #read csv comment
    danmu = pd.read_csv(filenames[i],engine='python',encoding='utf-8')
    #drop NA
    danmu.dropna(inplace=True)
    #drop "unnamed:0" useless column
    danmu.drop('Unnamed: 0',axis=1,inplace=True)
    #drop comments at same time by same id
    danmu_unique = danmu.drop_duplicates(['ID','danmu','data_unix']).reset_index()
    #join list to a string
    comments = " ".join(list(danmu_unique['danmu']))
    #jieba.cut -> tokenize
    danmu_fenci = list(jieba.cut(comments.strip(),cut_all=False))
    #remove stopwords
    refined_words = rm_stop(danmu_fenci,stopwords)
    #TF-IDF frequency analyse
    tags_tfidf = analyse.extract_tags(" ".join(refined_words),topK=20)
    #store each eposide's analysis results to dictionary
    HF_content[i+1]=tags_tfidf
    
```

```python
HF_content_df = pd.DataFrame(HF_content) #convert to df
HF_content_df.head()
```

### 24小时的弹幕频率分布

* unix time: 1970-01-01 00:00:00 Coordinated Universal Time (UTC)
*  = 1970-01-01 08:00:00 Beijing Time
* in seconds

```python
#put 12 episode comments to a dataframe
Danmu_season_df = pd.DataFrame()
for i in range(12):
    danmu = pd.read_csv(filenames[i],engine='python',encoding='utf-8')
    danmu.dropna()
    danmu.drop('Unnamed: 0',axis=1,inplace=True)
    danmu_unique = danmu.drop_duplicates(['ID','danmu','data_unix']).reset_index()
    Danmu_season_df = Danmu_season_df.append(danmu_unique,ignore_index=True)
```

```python
Danmu_season_df.append(danmu,ignore_index=True)
```

* process date time

```python
all_time = Danmu_season_df['data_unix']
all_time_array = np.array(all_time) + 8*60*60 #调整时差: UTC -> Beijing Time Unix
```

```python
all_time_array_24 = all_time_array%(24*60*60)/(60*60) #取余 一天=24*60*60 seconds, then convert unit from seconds->hour
```

```python
plt.figure(figsize=(11,7))
sns.distplot(all_time_array_24,bins=24,kde=False).set_title('Frequency within 24 hours')
sns.plt.xlim(0, 24)
sns.set_context("poster",font_scale=1.5)
```

### 年内的弹幕频率分布

```python
plt.figure(figsize=(11,7))
all_time = Danmu_season_df['data_unix']
all_time_array = np.array(all_time) + 8*60*60
all_time_array_365 = np.floor(all_time_array%(24*60*60*365)/(24*60*60)) #取余 一年=24*60*60*365 seconds, then convert unit from seconds->day
sns.distplot(all_time_array_365,bins=365,kde=False).set_title('Frequency within 365 Days')
sns.set_context("notebook",font_scale=1.5)
```

```python
from collections import Counter
C_d = Counter(all_time_array_365)
```
