```python
# Python3 的用法
#要去掉的元素放到 str.maketrans的第二个参数中
trantable = str.maketrans("","",string.punctuation)
def data_clean(text):
    text = text.lower()
    text_clean = text.translate(trantable)
    return text_clean
```

#### 4.3.4 词干提取与词性还原

```python
from nltk.stem import PorterStemmer 
from nltk import WordNetLemmatizer
from nltk import word_tokenize
```

```python
porter = PorterStemmer()
wnl = WordNetLemmatizer()
```

```python
print("stemmer: 'eating' ->%s"%(porter.stem('eating')))  #词干提取

print("lemmatization: 'feet' -> %s"%(wnl.lemmatize('feet',pos="n"))) #词形还原
print("lemmatization: 'feet' -> %s"%(wnl.lemmatize('feet',pos="v"))) #词形还原

print("lemmatization: 'ran' -> %s"%(wnl.lemmatize('ran',pos="n"))) #词形还原
print("lemmatization: 'ran' -> %s"%(wnl.lemmatize('ran',pos="v"))) #词形还原
print("lemmatization: 'Pumps' -> %s"%(wnl.lemmatize('Pumps',pos='n'))) #词形还原
print("lemmatization: 'pumps' -> %s"%(wnl.lemmatize('pumps',pos='n')))

# for lemmatization, 
# 1. If the first letter of the word is Capitalized, it will see it is a special word, it will not remove e.g. Pumps -> Pump
# 2. you need to specify pos(part of speech) for a word, e.g. verb or noun, otherwise it will not do the work. 

# So best practice is to 1. Lower case for proper words before Lemma(e.g. lower PUmp->pump, but not abbreviation like DCS-> dcs); 
#2. do lemma twice, first time by "n" for noun, second time by "v" for verb
# But still some combined word is hard to return to its real state (e.g. setpoints->setpoint), for example below, it will still
# return setpoints. So it is better to 3. test your keyword e.g. setpoints in lemma pos='n' before you proceed.
print("lemmatization: 'setpoints' -> %s"%(wnl.lemmatize('setpoints',pos='n')))

#--------------------------- Below is a summary example of the 3 points above------------------------------
text = 'This is a Demo Texts for NLP using NLTK. Full forms of NLTK is Natural Language Toolkits' #3.Be aware "toolkits" is a combined word, it is hard to return to toolkit by lemma
lower_text = text.lower() #1. lower case, but it will lower all ABBREVIATIONS e.g. DCS -> dcs, to prevent this, need user-defined function to lowercase
textlist1 = word_tokenize(lower_text)

print (textlist1)
lemma=[wnl.lemmatize(word,pos="n") for word in textlist1] #2.1 POS='n', but this will treat all word as noun
print(lemma)
lemma1=[wnl.lemmatize(word,pos="v") for word in lemma] #2.2 POS='v', but this will treat all word as verb
print (lemma1)

'''
To prevent the drawback of the summary example above, see what I did for ACM conditioning -> Archive->Data_cleaning_function_w_POS
In that function package, I user-defined the situation to lowercase the certain words (preventing lowercase the abbreviation)
Also I use POS_tag and user-defined rules, to separately lemma verb or noun word, so that it is a safer method compared to 
summary example above
'''
```

```python
def stem_tokens(tokens, stemmer = PorterStemmer()):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed

def tokenize(text):
    tokens = nltk.word_tokenize(text)
    #tokens = [word for word in tokens if word not in stopwords.words('english')]
    stems = stem_tokens(tokens)
    return ' '.join(stems)

def build_corpus(dataset):
    corpus = []
    for text in dataset:
        # without the decoding, there will be UnicodeDecodeError when tokenizing
        corpus.append(tokenize(text))  
    return corpus
```

#### 4.4.2 进一步处理，计算单词出现的tf-idf频率：sklearn TfidfVectorizer (会用就行)


* 什么是TF-IDF？
* TF: term frequency 具体单词在某篇文章中的出现次数 (比值越高说明这个词在该文章中越有代表性)
* $TF_{词i,文章j} = \frac{词i在文章j中的频率}{文章j中所有词的频率总和}$
* IDF: inversed document frequency -- 文章总数:带有该词汇i的文章数 (比值越高说明词汇i越能区分这类文章和别的文章)
* $IDF_{词i} = log \frac{所有文章的篇数总和}{带有词汇i的文章篇数总和}$

* Evaluation=TF-IDF=TF*IDF -- 综合考虑词汇i在该文章的代表性,和该类文章的代表性, 越高越说明该词汇代表性越高

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vec=TfidfVectorizer()

X_train_tfidf_vec=tfidf_vec.fit_transform(train_token)
X_test_tfidf_vec=tfidf_vec.transform(test_token)
```
