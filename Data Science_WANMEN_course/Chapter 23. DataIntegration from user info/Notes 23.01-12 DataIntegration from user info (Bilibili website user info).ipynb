### 2.1 正向编码方法：zlib.crc32


```python
# example of comment id: 8ce7a99b
```

```python
 mid  -> comment id   # bilibili uses zlib.crc32 to encode mid to comment id.
```

```python
import zlib
def crc32(bdata):
    h = zlib.crc32(bdata)
    x='%08X' % (h & 0xffffffff,)
    return(x.lower())
```

### 2.2 如何逆向解码？

We have comment id, our purpose is to get mid using [decoding of the comment id].

```python
mid                              ->     comment id    #crc32
mid [0,2 4000 0000]                      crc32(mid)
```

We decided to use Method 2

```python
# 注意对比顺序:
    # Order 1: 
        # For each row in 4 G file (240 million rows):
            # for each row in comment ids (360K rows)
                #match comment ids and mid
    # Order 2:
        # for each row in comment ids (360K rows):
            # for each row in 4 G file (240million rows)
                #match comment ids and mid
                
    # The two orders have the same complexity. But in practice, Order 1 takes 8 hours while order 2 takes 5 days, why?
        # - Because computer will put outer loop table into a RAM, then refer to it directly rather than read from hard drive. 
        # In order 1, 4G file has already been in RAM when first time loop, so later on, computer only needs to read 360K rows
        # inner loop table from hard drive; While in order 2, only 360K was read into RAM, computer needs to read 4G file from 
        # hard drive everytime when it goes through the innerloop.
        # So order 1 is faster than order 2 in this case, we choose order 1, see below
import re
with open(decode_file_name,'r') as f:
    for i,line in enumerate(f): # for each row in that 4G file (240Million rows)
        if (i-1)%10000==0: #for every 10000 rows, print something to give user a Feedback, telling you computer is still running
            print(i-1)
        if (i-1)%500000==0:
            gc.collect()
        if (i-1)%5000000==0: #for every 5 million rows, store the partial comparison results to csv, just in case computer is power outage.
            ids_pd.to_csv('ids_pd%d.csv'%(int((i-1)//1000000)))
        match=re.search(r",([a-zA-Z0-9]+)",line) # IF line matches the Regex format, it is one user comment id and can be used to search in comment_ids 
        # this step can be used to avoid selecting e.g. NaN, empty item in ucmid
        
        user_cmid = match.group(1)
        if user_cmid in comment_ids: #search if user comment id in comment_ids set
            ids_pd.loc[ids_pd['ID']==user_cmid,'orig_ID']=i-1
```
