### 2.3 扑腾扑腾翅膀

* 帧率 FPS - frame per second
* 如果不设置 FPS将会怎样
    * pygame_no_fps.py
* 设置后 FPS将会怎样
    * pygame_fps.py


Why FPS:
    
    > pygame is operated based on event (activity) created by user. (e.g. click mouse, click keyboard)
    
    > While True (game is not exit), computer will constantly look at event and refresh image
        > For example: 
                >User press down right arrow key ONCE -> this is one activity(event)
                >the time duration of this event might be 0.5s
                >consider the computer can check 100 times per second constantly.
                >Then this event created by user will be checked 50 times, i.e. computer will consider user "pressing down right arrow key" for 50 times.
                >Then flappy bird moves 50 times to the right by this event (by only one click), which is not what we want
                >FPS can limit how many times computer refresh display per second, to control how many times of "While True" loop is executed per second in the code.
