## 1. 贝叶斯思维

\begin{equation*}
P(A/B)=\frac{P(B/A)P(A)}{P(B)}\\
\end{equation*}


* 静态的思维：
    * From Static point of view, Bayes equation is just like conditional probability. (This was covered in Chapter 22)
    * 你头疼发烧，有多少概率是流感？
    * 你收到的短信里面有free, mobile等关键词，有多少概率是垃圾短信？
    


* 动态的思维：
    * From Dynamic point of view, Bayes equation is like the picture below.
    * 先验估计="Prior"=P(A); Likelihood=P(B|A), 后验估计="Posterior"=P(A|B)
    * 你怀疑你是流感，做了尿检检验，显示阳性，你对你的流感的怀疑增加了多少？-- iteration method
    * 你再进行查血检验，又是阳性，你对你患流感的怀疑变成了多少？

### 1.1 贝叶斯更新对违禁药物使用嫌疑的怀疑

第一次尿检
* 你之前很相信一位运动员，认为他不会使用违禁药品，或者使用违禁药品的概率非常低：P(使用禁药) = 0.001 --> P(A)
* 药检的likelihood是较为公正的，如果使用了，那么0.95会呈阳性 P(阳性／使用禁药)；但假阳性也比较高，即就算是清白的，也有0.1的概率会被显示成阳性 P(阳性／清白) --> P(阳性)=P(B)=P(B,A)+P(B,not A)=P(B|A)xP(A)+P(B|not A)xP(not A)
* 这位运动员第一次检验没有通过，你对他使用违禁药品的怀疑是多少？

P(使用禁药／阳性) = $\frac{P(阳性/使用禁药) \times P(使用禁药)}{P(阳性/使用禁药) \times P(使用禁药) + P(阳性/清白) \times P(清白)} $

= $\frac {0.95 \times 0.001}{(0.95 \times 0.001) + (0.10 \times 0.999)}$

$\approx$ 0.009




第二次尿检
* 你依然相信这位运动员是清白的 P(使用禁药) = 0.009 -> P(A) is updated based on the first exam above
* 这位运动员很不争气，在第二次尿检中，又没通过，相关的likelihood同第一次的likelihood
* 问题：第二次尿检没通过，你对他使用违禁药品的怀疑到了多少？

P(使用禁药／阳性) = $\frac{P(阳性/使用禁药) \times P(使用禁药)}{P(阳性/使用禁药) \times P(使用禁药) + P(阳性/清白) \times P(清白)} $

= $\frac {0.95 \times 0.009}{(0.95 \times 0.009) + (0.10 \times 0.991)}$

$\approx$ 0.079

第三次尿检
* 你依然相信他的清白 P(违规) = 0.079 -> P(A) is updated based on the second exam above
* 第三次尿检，他还是没通过
* 问这时你对他的怀疑有多少？

P(使用禁药／阳性) = $\frac{P(阳性/使用禁药) \times P(使用禁药)}{P(阳性/使用禁药) \times P(使用禁药) + P(阳性/清白) \times P(清白)} $

= $\frac {0.95 \times 0.079}{(0.95 \times 0.079) + (0.10 \times 0.921)}$

$\approx$ 0.45

Based on results above, P(A|B): 0.009->0.079->0.45...

ITERATION Method: 对于贝叶斯参数更新来说，不断增多的观测事件对参数更新有累计的效果，因为前一次获得数据后得到的**后验分布P(A|B)**，可以成为下一次获得新数据后进行贝叶斯参数更新的**先验分布P(A)**。

### 贝叶斯公式

P(A): 参数的先验分布，先验概率，先验估计，先验认识，先验信仰！
P(B/A): 在参数的先验概率成立的情况下，出现观测事件的概率
P(B): 任意情况下出现观测事件的概率

P(A/B): 后验概率，后验分布。出现了观测事件后，参数概率的分布

对于贝叶斯**参数更新**来说，不断增多的观测事件对参数更新有累计的效果，因为前一次获得数据后得到的<span style='color:blue'>**后验分布**</span>，可以成为下一次获得新数据后进行贝叶斯参数更新的<span style='color:blue'>**先验分布**</span>。

## Conjugate Distribution共轭分布

* Definition: If 在贝叶斯变化下，**参数的先验分布P(A)**和**参数后验分布P(A|B)**依然是**关于A的同一类概率分布**，那么the prior and posterior are then called conjugate distributions, and prior[参数的分布P(A)] is a conjugate prior for Likelihood的关于A的分布P(B|A)

* Definition in English: In Bayesian probability theory, if the posterior distributions p(θ | x) are in the same probability distribution family as the prior probability distribution p(θ), the prior and posterior are then called conjugate distributions, and the prior is called a conjugate prior for the likelihood function. 

* Why do we care?
    - If we know the Prior and its Likelihood is the conjugate prior distribution about A, --> Then we know A's Posterior is the same type of distribution as its Prior Distribution. 
    - Then when we do iteration(see lecture notes --"iteration method" above) for Bayes Model to update the A based on Likelihood, we don't need to refer to Bayes formula every time
    - Instead, we only need to update certain parameters since we already know its Posterior distribution.
    - For Example: as shown in the example 1 below, it is proved that Binomial Dist and Beta Dist are Conjugate prior Dist(see picture below) -> Then we know for a Bayes Model where Likelihood(P(data|theta)) is Binomial and Prior(P(theta)) is Beta Dist(i.e. Beta(a,b)) -> its Posterior(P(theta|data)) is Beta Dist as well. -> In such case, to get the explicit analytical solution of Posterior(P(theta|data)), we only need to get the relationship between a-> a' and b-> b' instead of relationship between Beta(a,b)-> Beta(a',b'). In other words, instead of iteration of Bayes formula to update Posterior, we only need to update key parameters, which will reduce the time of processing, especially when updating posterior many times, saying >100 times.

### Example 1. 二项分布(binomial distribution)和Beta分布是共轭分布

- If P(A)=beta distribution, and P(B|A)=binomial distribution -> then after calculation, P(A|B) is still shown as beta distribution (see prove below), --> Therefore we can say Beta and Binomial distribution are Conjugate Distribution 

#### Beta函数与Beta分布

* Beta Distribution
\begin{equation*}
P(x) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{Beta(\alpha,\beta)}
\end{equation*}
* 到底是什么含义？x是什么？$\alpha$是什么？$\beta$是什么？

- Say P(A)=P(x)

* 若新证据的likelihood是二项分布-> P(B|A)=
\begin{equation*}
P(s,f|x)=C_{s+f}^s x^s(1-x)^f
\end{equation*}

* 后验分布
\begin{equation*}
P(x|s,f) = \frac{P(s,f|x)P(x)}{\int P(s,f|x)P(x)dx}\\
= \frac{x^{s+\alpha-1}(1-x)^{f+\beta-1}}{Beta(s+\alpha,f+\beta)}\\
\end{equation*}

```python
def postor_dis(data,N):
    
# For the already prepared, I'm using Binomial's conj. prior.
# In Bayesian inference, the beta distribution is the conjugate 
# prior probability distribution for the Bernoulli, binomial, negative binomial and geometric distributions. 
    plt.figure(figsize=(6,5))
    plt.xlabel("$p$, probability of heads") 
    #    if k in [0, len(n_trials) - 1] else None
    #plt.setp(sx.get_yticklabels(), visible=False)
    heads = data[:N].sum()
    y = dist.pdf(x, 1 + heads, 1 + N - heads)
    plt.plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)

    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)


    plt.suptitle("Bayesian updating of posterior probabilities",
                 y=1.02,
                 fontsize=14)

    plt.tight_layout()
```

```python
%matplotlib inline
from IPython.core.pylabtools import figsize
import numpy as np
import matplotlib.pyplot as plt
figsize(11, 9)

import scipy.stats as stats

dist = stats.beta
n_trials = [0, 1, 2, 3, 4, 5, 8, 15, 50, 500]
data = stats.bernoulli.rvs(0.5, size=n_trials[-1])
x = np.linspace(0, 1, 100)

# For the already prepared, I'm using Binomial's conj. prior.
# In Bayesian inference, the beta distribution is the conjugate 
# prior probability distribution for the Bernoulli, binomial, negative binomial and geometric distributions. 
for k, N in enumerate(n_trials):
    sx = plt.subplot(len(n_trials) / 2, 2, k + 1)
    plt.xlabel("$p$, probability of heads") \
        if k in [0, len(n_trials) - 1] else None
    plt.setp(sx.get_yticklabels(), visible=False)
    heads = data[:N].sum()
    y = dist.pdf(x, 1 + heads, 1 + N - heads)
    plt.plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)

    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)


plt.suptitle("Bayesian updating of posterior probabilities",
             y=1.02,
             fontsize=14)

plt.tight_layout()
```

## 证明正态分布的后验分布依然是正态分布

* $\sigma$固定，只更新$\mu$

* 先验分布
\begin{equation*}
p(\mu) \propto exp(-\frac{1}{2\sigma_0^2}(\mu-\mu_0)^2) \\
\propto  N(\mu|\mu_0,\sigma_0^2)
\end{equation*}

* likelihood出现观测事件的概率
\begin{equation*}
p(D|\mu) \propto exp(-\frac{1}{2\sigma^2}(\sum(x_i-\mu)^2) \\
p(D|\mu) \propto exp(-\frac{1}{2\sigma^2}(\sum(x_i-\bar{x}+\bar{x}-\mu)^2) \\
p(D|\mu) \propto exp(-\frac{n}{2\sigma^2}(\bar{x}-\mu)^2) \\
\end{equation*}

<img src='Norm2.jpg' height=300 >

* 后验分布
\begin{equation*}
p(\mu|D)\propto p(D|\mu)p(\mu|\mu_0,\sigma_0^2)\\
\propto exp[-\frac{1}{2\sigma^2}\sum (x_i-\mu)^2] \times exp[-\frac{1}{2\sigma_0^2}(\mu-\mu_0)^2]\\
=exp[-\frac{1}{2\sigma^2}\sum (x_i^2+\mu^2-2\mu x_i)^2-\frac{1}{2\sigma_0^2}(\mu^2+\mu_0^2-2\mu_0\mu)^2]\\
=exp[-\frac{\mu^2}{2}(\frac{1}{\sigma_0^2}+\frac{n}{\sigma^2})+
\mu (\frac{\mu_0}{\sigma_0^2}+\frac{\sum_i x_i}{\sigma^2})-
(\frac{\mu_0^2}{2\sigma_0^2}+\frac{\sum_i x_i^2}{2\sigma^2})]\\
\propto exp[-\frac{1}{2\sigma_n^2}(\mu^2-2\mu\mu_n+\mu_n^2)]
\end{equation*}

See proof in picture above, where \begin{equation*}u^* is \mu_n below\end{equation*} , same for sigma

\begin{equation*}
\sigma_n^2 = \frac{\sigma^2\sigma_0^2}{n\sigma_0^2+\sigma^2}
=\frac{1}{\frac{n}{\sigma^2}+\frac{1}{\sigma_0^2}}
\end{equation*}

\begin{equation*}
\mu_n=\frac{\sigma^2}{n\sigma_0^2+\sigma^2}\mu_0+\frac{n\sigma_0^2}{n\sigma_0^2+\sigma^2}\bar{x}
\end{equation*}

* 总结起来，就是

\begin{equation*}
x_i|\mu \sim N(\mu,\sigma^2) i.i.d \ and \ \mu \sim N(\mu_0,\sigma_0^2)\\
then\\
\mu|x_1,x_2,x_3,...,x_n \sim N(\frac{\sigma_0^2}{\frac{\sigma^2}{n} +\sigma_0^2}\bar{x}+\frac{\sigma^2}{\frac{\sigma^2}{n}+\sigma_0^2}\mu_0,(\frac{1}{\sigma_0^2}+\frac{n}{\sigma^2})^{-1})
\end{equation*}

```python
%matplotlib inline
from IPython.core.pylabtools import figsize
import numpy as np
import matplotlib.pyplot as plt
figsize(11, 9)

import scipy.stats as stats

dist = stats.norm
n_trials = [1, 5, 10,20, 50,100, 200, 300,400,500]
#以下为先验分布的参数
mu_0=0
sigma_0=1
#以下为真实分布
data = stats.norm.rvs(loc=1.4,scale=5, size=n_trials[-1])
sigma=5

x = np.linspace(-1, 4, 100)

# In Bayesian inference 正太分布的共轭分布是正太分布
for k, N in enumerate(n_trials):
    sx = plt.subplot(len(n_trials) / 2, 2, k + 1)
    plt.xlabel("$\mu$ ") \
        if k in [0, len(n_trials) - 1] else None
    plt.setp(sx.get_yticklabels(), visible=False)
    #用贝叶斯方法更新参数分布
    x_bar = np.mean(data[:N])
    n=N
    mu_n = sigma_0**2/(sigma**2/n+sigma_0**2)*x_bar + sigma**2/(sigma**2/n+sigma_0**2)*mu_0
    sigma_n = np.sqrt(1/(1/sigma_0**2+n/sigma**2))
    
    y = dist.pdf(x, loc=mu_n, scale=sigma_n)
    plt.plot(x, y, label="numbers of observation %d"%n)
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(1.4, 0, 4, color="k", linestyles="--", lw=1)

    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)


plt.suptitle("Bayesian updating of posterior probabilities",
             y=1.02,
             fontsize=14)

plt.tight_layout()
```
